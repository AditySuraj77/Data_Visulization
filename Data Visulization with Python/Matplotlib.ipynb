```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1,11)
x
```

```python
y = x*3
```

```python
y
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y)
plt.title('Graphs')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
plt.plot(x, y, color='r', linestyle=':', linewidth='5')
plt.title('Graphs')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.grid(True)
plt.show()
```

# Plot Two line in One graph

```python
y1 = x*10
y2 = x*20
```

```python
plt.plot(x, y1, color='r', linestyle=':', linewidth='2')
plt.plot(x, y2, color='b', linestyle=':', linewidth='2')
plt.title('Graphs')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.grid(True)
plt.show()
```

# Ploting Subplot

```python
plt.subplot(1,2,1)
plt.plot(x, y1, color='r', linewidth='2')

plt.title('Graphs')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.grid(True)


plt.subplot(1,2,2)
plt.plot(x, y2, color='b', linewidth='2')

plt.title('Graphs')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.grid(True)
plt.show()
```

# Bar Plot

```python
student= {"Kl":54, "Virat":60, "Ro-Hit":56}
```

```python
name = list(student.keys())
values = list(student.values())
```

```python
plt.bar(name, values)
plt.title('Bar Graphs')
plt.xlabel('Names')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```

```python
plt.barh(name, values, color='g')
plt.title('Bar Graphs')
plt.xlabel('Names')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```
