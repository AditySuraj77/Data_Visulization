```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = pd.read_csv('cereal.csv')[0:10]
iris
```

```python
sns.boxplot(x='calories', y='sugars', data=iris)
plt.show()
```

```python
sns.boxplot(x='calories', y='sugars', data=iris, palette='Set1')
plt.show()
```

```python
sns.boxplot(x='calories', y='sugars', data=iris, linewidth=3)
plt.show()
```

```python
sns.boxplot(x='calories', y='sugars', data=iris,color='orange')
plt.show()
```

```python
sns.boxplot(x='calories', y='sugars', data=iris,hue='weight')
plt.show()
```

```python
sns.pairplot(iris, hue='calories')
plt.plot()
```
