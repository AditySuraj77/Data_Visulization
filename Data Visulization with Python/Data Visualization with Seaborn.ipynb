```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
trn = pd.read_csv('train.csv')
```

```python
trn.head()
```

```python
trn1 = trn[0:10]
```

```python
sns.lineplot(x='Age', y='Sex', data=trn1)
plt.show()
```

```python
sns.lineplot(x='Age', y='Sex', data=trn1, hue='Pclass')
plt.show()
```

```python
sns.lineplot(x='Age', y='Sex', data=trn1, hue='Pclass', style='Embarked')
plt.show()
```

```python
sns.lineplot(x='Age', y='Sex', data=trn1, hue='Pclass', style='Embarked', markers=True)
plt.show()
```

# Barplot

```python
sns.set(style='whitegrid')
pokemon = pd.read_csv('pokemon.csv')
```

```python
pokemon.head()
```

```python
sns.barplot(x='Legendary', y='Speed', data=pokemon)
plt.show()
```

```python
sns.barplot(x='Legendary', y='Defense', data=pokemon)
plt.show()
```

```python
sns.barplot(x='Legendary', y='Speed', data=pokemon, hue='Generation')
plt.show()
```

```python
sns.barplot(x='Legendary', y='Speed', data=pokemon , palette='husl')
plt.show()
```
