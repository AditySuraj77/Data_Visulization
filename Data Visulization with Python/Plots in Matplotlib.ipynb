```python
import matplotlib.pyplot as plt
```

# Scatter plot

```python
x= [10,20,30,40,50,60,70,80,90]
a = [1,5,7,89,5,35,5,4,4]
b = [15,50,7,8,5,3,5,54,4]
```

```python
plt.scatter(x,a)
plt.show()
```

```python
plt.scatter(x,a, marker="<", c= 'r', s=100)
plt.scatter(x,b, marker="+", c= 'b', s=100)
plt.show()
```

# Subplot of Scatter plot

```python
plt.subplot(1,2,1)
plt.scatter(x,a, marker="<", c= 'r', s=100)

plt.subplot(1,2,2)
plt.scatter(x,b, marker="+", c= 'b', s=100)

plt.show()
```

# Histogram

```python
data = [10,22,13,4,5,5,4,13,26,10]
```

```python
plt.hist(data)
plt.show()
```

```python
plt.hist(data, color='r', bins=6)
plt.show()
```

# Boxplot

```python
one = [1,2,3,4,5,5,4,3,2,1]
Two = [2,4,6,8,10,10,8,6,4,2]
thre= [9,8,7,6,5,4,3,2,1]
```

```python
data = list([one, Two, thre ])
```

```python
plt.boxplot(data)
plt.show()
```

# ViolinPlot

```python
plt.violinplot(data, showmeans=True)
plt.show()
```

# Pie Chart

```python
fruit = ['Apple', 'Orange', 'PineApple', 'Banana', 'Graphes']
Quantity = [25,15,20, 30, 10]
```

```python
plt.pie(Quantity, labels=fruit)
plt.show()
```

```python
plt.pie(Quantity, labels=fruit, autopct='%0.1f%%', colors=['yellow', 'green', 'red', 'blue', 'white'])
plt.show()
```

# Dought Nut - Chart

```python
plt.pie(Quantity, labels=fruit, radius=2)
plt.pie([10], colors=['w'], radius=1)
plt.show()
```
